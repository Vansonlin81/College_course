# 產生樣本點

產生線性樣本點

```python
sample1 = np.arange(0., 5., 0.2)
sample2 = sample1.copy()
np.random.shuffle(sample2)
sample3 = sample1.reshape(5, 5)
sample4 = sample1.reshape(25)
# 將元素型態改為「文字」，並將之印出
sample5 = sample1.astype('unicode')

print(sample1, '\n')
print(sample2)
print(sample3)
print(sample4)
print(sample5)
# 因為電腦有誤差, 十進位的有限小數轉成二進位時會變成無限小數再因為電腦記憶體有限會被截斷, 所以小數不叫小數, 叫浮點數.
```

產生「等差」樣本點 \
產生「等比」樣本點

```python
sample1_1 = np.linspace(0., 5., 15)
sample1_2 = np.logspace(0., 5., 15)

print(sample1_1)
print(sample1_2)
```

產生「整數亂數」樣本點 \
產生「浮點數數亂數」樣本點

```python
sample2_1 = np.random.randint(0, 10, 100) # 0~10 choose 100 integer
sample2_2 = np.random.rand(2, 10) # 產生 [0, 1) 之間的浮點亂數 2x10 個 （其它範圍亂數：以 r [0, 1) + b 製作之）
sample2_3 = 3*np.random.rand(2, 10) + (-2) # 產生 [-2, 1) 之間的浮點亂數 2x10 個

print(sample2_1)
print(sample2_2)
print(sample2_3)
```

產生「標準常態分佈N(0, 1)」樣本點 \
產生「⼀般常態分佈」樣本點

```python
sample3_1 = np.random.randn(3, 5)
sample3_2 = np.random.normal(10, 2, size = (3, 5))

print(sample3_1)
print(sample3_2)
```

切片運算(Slicing)

```python
# 1-dim
slice1 = np.random.randint(4, size = 20)

print(slice1)
print(slice1[:5])

# 2-dim
slice2 = np.random.randint(4, size = (3, 5))

print(slice2)
print(slice2[:3, :3])
```

統計樣本資訊.

```python
stat1 = np.arange(0, 100)

print(stat1)

print(stat1.min())
print(stat1.max())
print(stat1.sum())
print(stat1.mean())
print(stat1.std())
```

```python
sample4_1 = np.array([1, 2, 3, 4])
sample4_2 = np.array([5, 6, 7, 8])
sample4_3 = np.concatenate((sample4_1, sample4_2))
sample4_4 = np.concatenate((sample4_2, sample4_1))
sample4_5 = np.concatenate((sample4_2, sample4_1)).reshape(2, 4)
sample4_6 = np.concatenate((sample4_2, sample4_1)).reshape(2, 4).T

print(sample4_1)
print(sample4_2)
print(sample4_3)
print(sample4_4)
print(sample4_5)
print(sample4_6)
```
