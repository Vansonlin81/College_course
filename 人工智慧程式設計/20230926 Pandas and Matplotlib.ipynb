# Pandas Introduce

```python
import pandas as pd
Rows = ['Row1', 'Row2', 'Row3']
Columns = ['Column1', 'Column2']

# by list (Row-based).
dfList = pd.DataFrame([[1, 2], [3, 4], [5, 6]], index = Rows, columns = Columns)
print(dfList)

print('-'*50)

# by Dict (Column-based).
dfDict = pd.DataFrame({'Column1':[1, 3, 5], 'Column2':[2, 4, 6]}, index = Rows)
print(dfDict)
```

```python
# 缺失資料NaN 補齊方法！
dfCarSales_csv = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/人工智慧程式設計/CarSales.csv')
print(dfCarSales_csv)
print('-'*50)
print(dfCarSales_csv['Country'].mode())
print('-'*50)
print(dfCarSales_csv['Age'].median())
print('-'*50)
print(dfCarSales_csv['Salary'].mean())
```

```python
# 取得資料集的‘摘要描述’
dfCarSales_csv.describe(include = 'all')
```

```python
# 取得特定欄列
print(dfCarSales_csv.iloc[:, 1:3]) # 取第一第二個 column
print('-'*50)
print(dfCarSales_csv.loc[:, ['Country', 'ToBuy']])
```

```python
# 加入條件
# 先將條件儲存成變數，再將條件塞入[]
the_condition = (dfCarSales_csv['Age'] != 'NaN')
dfCarSales_csv[the_condition]
```

```python
# 轉成 Numpy array
ary1 = dfCarSales_csv.to_numpy()
print(ary1)
print('-'*50)
ary2 = dfCarSales_csv.values
print(ary2)
```

## Line Charts

```python
# only y-axis
plt.plot([2, 4, 6, 8])
plt.show()
```

```python
# x-axis and y-axis
plt.plot([1, 3, 5, 7, 9], [2, 4, 6, 8, 10])
plt.show()
```

多組資料

```python
import numpy as np
X = np.arange(0., 5., 0.2)
plt.plot(X, X, 'r--')
plt.plot(X, X**2, 'c^')
plt.plot(X, X**3, 'g-.')
```

```python
from matplotlib.font_manager import FontProperties
myfont = FontProperties(fname = r'/content/drive/MyDrive/Colab Notebooks/人工智慧程式設計/fonts/KAIU.TTF')
```

```python
X = np.arange(0., 5., 0.2)
plt.title('X, X^2, X^3 線圖', fontproperties = myfont)
plt.xlabel('x-axis', fontproperties = myfont)
plt.ylabel('y-axis', fontproperties = myfont)

plt.plot(X, X, 'r--', label = 'X')
plt.plot(X, X**2, 'bs', label = 'X**2')
plt.plot(X, X**3, 'g^', label = 'X**3')

plt.legend(loc = 'best') # 圖例的位置 'center', 'upper left', 'lower'
plt.show()
```

## Scatter Charts

```python
# size of sample data
n = 150

# Group 1 Data around (-1, 2) ~ND(0, 1)
X1 = np.random.normal(-1, 1, n)
Y1 = np.random.normal(2, 1, n)

# Group 1 Data around (2, -1) ~ND(0, 1)
X2 = np.random.normal(2, 1, n)
Y2 = np.random.normal(-1, 1, n)

# plot
plt.scatter(X1, Y1, s = 75, c = 'red', marker = '+')
plt.scatter(X2, Y2, s = 75, c = 'blue', marker = 'o')

plt.show()
```
