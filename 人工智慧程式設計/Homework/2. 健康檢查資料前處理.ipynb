```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/人工智慧程式設計/Homework/HealthCheck.csv')
print(dataset)
```

檢查是否有缺失資料, 再算每列有幾個, 再看看整個dataframe總共有幾個缺失資料

```python
sum(dataset.isnull().sum())
```

切分自變數、應變數

```python
X = dataset.iloc[:, :-1].values
Y = dataset.iloc[:, 3].values
```

```python
X, Y
```

處理缺失資料

```python
from sklearn.impute import SimpleImputer
# 先做 Height_cm

imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean') # 將NaN值用平均值代替.
imputer = imputer.fit(X[:, 1:2])
X[:, 1:2] = imputer.transform(X[:, 1:2])

# 再做 Weight_cm

imputer = SimpleImputer(missing_values = np.nan, strategy = 'median') # 將NaN值用平均值代替.
imputer = imputer.fit(X[:, 2:3])
X[:, 2:3] = imputer.transform(X[:, 2:3])
```

```python
X
```

類別資料數位化.

```python
# 使用標籤編碼器，將應變數 Y 數位化
from sklearn.preprocessing import LabelEncoder

labelEncoder = LabelEncoder()
Y = labelEncoder.fit_transform(Y).astype("float64")
```

```python
print(Y)
```

```python
# 使用獨熱編碼器(one-hot encoding)，將自變數 X 數位化
ary_dummies = pd.get_dummies(X[:, 0]).values
X = np.concatenate((ary_dummies, X[:, 1:3]), axis=1).astype("float64")
```

```python
print(X)
```

切割訓練集和測試集

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
X_train, X_test, Y_train, Y_test
```

特徵縮放

```python
from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler().fit(X_train)
X_train = sc_X.transform(X_train)
X_test = sc_X.transform(X_test)
```
