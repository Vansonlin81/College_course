```python
from matplotlib.font_manager import FontProperties
myfont = FontProperties(fname = r'/content/drive/MyDrive/Colab Notebooks/人工智慧程式設計/fonts/KAIU.TTF')
```

```python
temp = dfweather['氣溫(℃)']
x = np.arange(1., 31., 1)
plt.plot(x, temp, label = 'tempreture')

plt.title('2023-06 溫度', fontproperties = myfont)
plt.xlabel('日期', fontproperties = myfont)
plt.ylabel('氣溫', fontproperties = myfont)
plt.legend(loc = 'best')
plt.show()
```

# Real Data

```python
dfweather = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/人工智慧程式設計/Homework/C0AC80-2023-06.csv')
print(dfweather)
```

```python
print(dfweather.values)
```

```python
dfweather.describe(include = 'all')
```

```python
dfweather.columns
```

```python
# .astype(float) 更改資料格式為"浮動數"
print('平均溫度', dfweather['氣溫(℃)'][1:].astype(float).mean())
print('-'*50)
print('最高濕度', dfweather['相對溼度(%)'][1:].astype(float).max())
print('-'*50)
print('最低氣壓', dfweather['測站最低氣壓(hPa)'][1:].astype(float).min())
print('-'*50)
print('總雨量：', dfweather['降水量(mm)'][1:].astype(float).sum())
```

```python
from matplotlib.font_manager import FontProperties
myfont = FontProperties(fname = r'/content/drive/MyDrive/Colab Notebooks/人工智慧程式設計/fonts/KAIU.TTF')
```

```python
temp = dfweather['氣溫(℃)'][1:].astype(float)
x = np.arange(1., 31., 1)
plt.plot(x, temp, label = 'tempreture')

plt.title('2023-06 溫度', fontproperties = myfont)
plt.xlabel('日期', fontproperties = myfont)
plt.ylabel('氣溫', fontproperties = myfont)
plt.legend(loc = 'best')
plt.show()
```
