1. 使用 pandas 讀取 CSV 檔案，並將讀入的資料集，顯示在螢幕上（三分）。
2. 使用 numpy 計算每一科的平均分數，並用 print，將各科平均分數印出來（四分）。
3. 使用 matplotlib.pyplot 繪製一個 line chart，展示各科目的平均分數。需包含圖形的 Title、xlabel、與 ylabel（三分）。

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
sc_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/人工智慧程式設計/Homework/Student_Score.csv')
```

```python
print(sc_df)
```

```python
print('Math average',sc_df.iloc[:, 1:2].mean())
print('English average',sc_df.iloc[:, 2:3].mean())
print('Science average',sc_df.iloc[:, 3:4].mean())
```

```python
plt.plot(['Math', 'English', 'Science'], [sc_df.iloc[:, 1:2].mean(), sc_df.iloc[:, 2:3].mean(), sc_df.iloc[:, 3:4].mean()], '.-')
plt.title('Average Scores in Each Subject')
plt.xlabel('Subjects')
plt.ylabel('Average Score')
```
