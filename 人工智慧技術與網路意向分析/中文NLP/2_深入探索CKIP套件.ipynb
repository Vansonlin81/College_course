```python
from ckip_transformers.nlp import CkipWordSegmenter, CkipPosTagger, CkipNerChunker

# 定義資料
text = [ "等到潮水退了，就知道誰沒穿褲子。",
      "數大便是美",
      "傅達仁今將執行安樂死，卻突然爆出自己20年前遭緯來體育台封殺，他不懂自己哪裡得罪到電視台。",
      "美國參議院針對今天總統布什所提名的勞工部長趙小蘭展開認可聽證會，預料她將會很順利通過參議院支持，成為該國有史以來第一位的華裔女性內閣成員。",
      "空白 也是可以的～",]
```

```python
# Initialize drivers
# tokenizer
# device=0: use GPU ; device=-1: without GPU
# model : 使用特定模型
ws_driver = CkipWordSegmenter(device=-1, model='bert-base')  # albert-base # 'bert-base'

ws_result = ws_driver(text)
```

```python
print(ws_result)
```

```python
# 詞性分析
# device=0: use GPU ; device=-1: without GPU
# model : 使用特定模型
pos_driver = CkipPosTagger(device=-1, model='albert-base') # ALBERT-base # 'bert-base'
pos_result = pos_driver(text)
print(pos_result)
```

```python
# Pack word segmentation and part-of-speech results
def pack_ws_pos_sentece(sentence_ws, sentence_pos):
   #assert len(sentence_ws) == len(sentence_pos)
   res = []
   for word_ws, word_pos in zip(sentence_ws, sentence_pos):
      res.append(f"{word_ws}({word_pos})")
   return "\u3000".join(res)
```

```python
# 隨便印一下結果，看看對應關係
print(pack_ws_pos_sentece(ws_result, pos_result))
```

```python
# Show results
for sentence, sentence_ws, sentence_pos in zip(text, ws_result, pos_result):
   print(pack_ws_pos_sentece(sentence_ws, sentence_pos))
   print()
```

```python
# 名稱實體辨識 NER
# device=0: use GPU ; device=-1: without GPU
# model : 使用特定模型
ner_driver = CkipNerChunker(device=-1, model='albert-base')  # albert-base # 'bert-base'
ner_result = ner_driver(text)
print(ner_result)
```
