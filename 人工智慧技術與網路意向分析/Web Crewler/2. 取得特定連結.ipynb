### Quiz 3 -- 用這個連結來練習，練習取出至少一篇新聞標題與內文、並分享你的爬蟲心得。

連結(3)：https://search.ltn.com.tw/list?keyword=懲戒權

````python
import requests

# 設定網頁 URL，就去抓
url = "https://search.ltn.com.tw/list?keyword=懲戒權"
page = requests.get(url)
````

### 步驟一：取得 (a) 標題 (b) 新聞連結

````python
# 先決定要抓的資料在哪裡？用什麼方式抓它們？

from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content, "html.parser")

#page in html structure
print(soup.prettify())

# remove all tags only the text you can see on the page not from the source code
#print(soup.get_text())
# 先看看這頁是些什麼東西
````

````python
# 試試不一樣的路徑寫法

# 處理完第一頁之後,我們獲得了哪些資料？
# 標題
# 連結
# 日期不可靠，故不在這裡抓
# 其實有部分新聞內容，但不全

#########################################################################
#news_path = "#ec > section > div.page-name > ul > li:nth-child(1) > div > a.tit"  # copy selector
news_path = "#ec > section > div.page-name > ul" # 取X_Path(好像是...)

#listobs = soup.find_all(news_path)
listobs = soup.select(news_path)
print(listobs)
````

````python
# 處理完第一頁之後,我們獲得了哪些資料？
# 標題
# 連結
# 日期不可靠，故不在這裡抓
# 其實有部分新聞內容，但不全


#########################################################################
# unique tag  '<ul class="list boxTitle" data-desc="列表">'
menuobj = soup.find("ul", {"class": "list boxTitle"})

#print(menuobj)
# 然後找出它後面的 list objects...
listobjs = menuobj.select('li')

# 開個 list 來存結果
my_list_news = []

for i, a_news in enumerate(listobjs):
    print(f'第{i}篇新聞：') # , a_news.get_text())
    # 因為標題和連結都藏在 <a> tag　裡，故要先選到<a>
    the_obj = a_news.select_one('a')
    # 找到標題
    the_title = the_obj.get('title')
    print('標題：', the_title)
    the_link = the_obj.get('href')
    print('連結：', the_link)

    my_list_news.append( { 'title': the_title, 'link' : the_link } )
````

步驟二：使用新聞連結，取得 (c) 新聞內文

````python
# 上面的新聞，有２０筆，已經存到了　my_list_news　對吧？
# 再來還要抓內文，要怎麼抓呢？
# #ltnRWD > div.content > section > div:nth-child(16) > div.text.boxTitle.boxText

import time

# 還是要再跑一次迴圈
for a_news in my_list_news:
  the_url = a_news['link']
  the_page = requests.get(the_url)
  soup_content = BeautifulSoup(the_page.content, "html.parser")

  content_obj = soup_content.find("div", {"class": "text boxTitle boxText"})
  #print(content_obj)
  the_news = content_obj.find_all('p', {'class': None})
  #print(the_news)
  the_content = '\n'.join( [news.text for news in the_news] )
  #print(the_content)
  a_news['content'] = the_content

  # 一定要睡一下，自由時報BAN人粉恐怖～～～
  time.sleep(5)
````

````python
# 印出蒐集後的結果
#print(my_list_news)

for mynews in my_list_news:
  print( mynews )
````

````python
import requests
from bs4 import BeautifulSoup

url = 'https://search.ltn.com.tw/list'
params = {'keyword': '懲戒權'}

response = requests.get(url, params=params)
soup = BeautifulSoup(response.text, 'html.parser')
````

````python
for news in soup.select('.searchlist li'):
    title = news.select_one('a').text.strip()
    link = news.select_one('a')['href']
    date = news.select_one('.time').text.strip()
    print(f'{date} - {title} ({link})')
````
