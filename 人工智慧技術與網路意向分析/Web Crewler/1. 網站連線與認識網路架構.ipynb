### 使用 BeautifulSoup 套件

網頁(1)：http://140.119.164.107/matytest/w5_example1.html



**常用函式／資料：**
```
get_text()
obj.text                ：取出 tag 中的文字內容
```
```
find(str_tag, ...)      ：找出下層中特定的 tag
```
```
find_all(str_tag, ...)
findAll(str_tag, ...)   ：找出下層中全部有此特定的 tag; attr 可指定其他字串，或使用 re; limit 可限制數量.
```
```
select(strselector, ...)：用 selector 取出下層特定位置內容
```




````python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content, "html.parser")

print(soup.prettify())#page in html structure
````

````python
# remove all tags only the text you can see on the page not from the source code
print(soup.get_text())
````

````python
print(soup.text) # text == get_text()
````

````python

#soup.title
# <title>The Dormouse's story</title>
print(f"title: {soup.title}")

#soup.title.name
# <title> tag name
print(f"title name: {soup.title.name}")

#soup.title.text# equals to print(soup.title.string)
# string inside <title> tag
print(f"title string: {soup.title.string}")

#use find 
print(soup.find("title"))

print(soup.find("h1"))#print(soup.find.h1) note: doesn't work this way

print (soup.find(id="link3"))


div = soup.find("div")

print(div)
if div:    #因為可能會沒找到 tag...
    print(div.select("a"))
````

````python

print(soup.find("title").text)
print(soup.find("title").string)

#soup.title.parent.name
# <title> tag's parent tag
print(f"title parent's name': {soup.title.parent.name}")

#soup.p
# everything inside p tag
print(f"pargraph: {soup.p}")

#soup.a
# find first hyper link
print(f"hyper link: {soup.a}")

#soup.find_all('a')
# find all hyper links
print(f"hyper links: {soup.find_all('a')}")

find_all = soup.find_all("h1",id="abc", limit = 2)
print (f"hyper link:{find_all}")
#print nothing if the find_all found nothing

for link in soup.find_all('a'):
    print(link.get('href'))

````

````python
#另一種方便的寫法：把它當作 dict object
print(f"hyper link: {soup.a['href']}")
````

````python
# 設定網頁 URL，就去抓
url_sub = soup.a['href']
page_sub = requests.get(url_sub)

print(page_sub.content)
````

````python
# 目標：確定國家位置，並用 soup 找出來
# 第一種寫法；但要先確認沒有其他資料一樣用 <li> tag
get_list = soup.find_all("li")
print(get_list)

#顯然結果怪怪的
````

````python
# 確定國家位置，並用 soup 找出來
# 另一種寫法；國家那區塊的範圍，外面好像有一個大 ol~
data_loc = "body > content-type: > ol > li:nth-child(1)"
get_list = soup.find("ol")

print(get_list)

#然後抓國家
w_obj = get_list.select('li')
print(w_obj)

for obj in w_obj:
    print(obj.get_text())
````

````python
# 第三種寫法
# Using selector
liobjs = soup.select(" ol > li")
#print(liobjs)
for obj in liobjs:
    print(obj.text)
````

````python
# try soup.select(id)
select_obj = soup.select('#link3')
print(select_obj)
````

### Quiz 2 -- 請用程式抓出『網頁(2)』中的分數資料，並計算平均。

網頁(2)：http://140.119.164.107/matytest/w5_example2.html

````python
import requests

# 設定網頁 URL，就去抓
url = "http://140.119.164.107/matytest/w5_example2.html"
page = requests.get(url)

#print(page.content)

#then?
## body > content-type: > table > tbody > tr:nth-child(2) > td:nth-child(3)
soup = BeautifulSoup(page.content, 'html.parser')

rows = soup.findAll('tr')
#print(rows)
total = []
for row in rows:
    if row:
        td_list = row.findAll('td')
        if td_list:
            #print(td_list[-1].text)
            try:
                total.append(int(td_list[-1].text))
            except ValueError:
                continue

print( sum(total) / len(total) )
````
