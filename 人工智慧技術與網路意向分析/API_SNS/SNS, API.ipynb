```python
# 範例：搜尋字詞

import tweepy

# 請自行代入自己申請的 token
bearer_token = "AAAAAAAAAAAAAAAAAAAAAChybQEAAAAAgRp0W%2BWLZtrTNqID8ICuuRKx3co%3DIJcOlZDATDNrC4j5SMxRbVhX3vqMnAuAovpNIcYmo6U5AmCap8"

client = tweepy.Client(bearer_token)

# Search Recent Tweets

# This endpoint/method returns Tweets from the last seven days

str_keyword = "烏克蘭"

# 某些欄位要加了才會取得
response = client.search_recent_tweets(str_keyword, tweet_fields=["created_at", "attachments","source"])

#, user_fields = ["name", "username", "location", "verified", "description"]
# The method returns a Response object, a named tuple with data, includes,
# errors, and meta fields
print(response.meta)

# In this case, the data field of the Response returned is a list of Tweet
# objects
tweets = response.data

print('總共搜尋到: ', len(tweets))
# Each Tweet object has default ID and text fields
for tweet in tweets:
    print('-'*80)
    #print(tweet.data)
    print(tweet.id)
    print(tweet.source)
    print(tweet.created_at)
    print(tweet.attachments)
    print(tweet.text)


# By default, this endpoint/method returns 10 results
# You can retrieve up to 100 Tweets by specifying max_results
response = client.search_recent_tweets(str_keyword, max_results=100)


```
