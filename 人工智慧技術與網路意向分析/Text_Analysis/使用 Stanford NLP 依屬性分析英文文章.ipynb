````python
#開始使用 stanfordcorenlp
from stanfordcorenlp import StanfordCoreNLP

nlp = StanfordCoreNLP('./stanford-corenlp-full-2018-02-27', timeout=10000)

# 分字/辭
test_str1 = 'The brown fox is quick and he is jumping over the lazy dog'
tokens = nlp.word_tokenize(test_str1)
print('====> word_tokenize:', tokens)

# 詞性標記
print('====> pos_tag:', nlp.pos_tag(test_str1))

# 描繪為樹狀形式
print('====> parse:', nlp.parse(test_str1))

# 解析為帶有樹狀結構的資訊
result = list(nlp.dependency_parse(test_str1))
print(result)
````

補充說明：常用的詞性標記對照

```
ROOT：    待處理的語句（一整句，無指涉任何詞）
IP：      從屬子句
NP：      名詞子句
VP：      動詞子句
PP：      介詞子句
ADVP：    副詞子句
ADJP：    形容詞子句
DNP：     由「的」連接構成的詞
CP：      由「的」構成的修飾短句
DP：      限定詞子句
QP：      量詞子句
NN：      常用名詞
NR：      專有名詞
NT：      時間名詞
PN：      代詞
VV：      動詞
VC：      是
CC：      連接詞
```
