以下是處理中文區塊

這次用這篇：https://ent.ltn.com.tw/news/breakingnews/4238532

```python
# 選一篇中文新聞來玩玩
Title_LibTime = '網湧入道賀奪奧斯卡 楊繡惠傻眼'
Content_LibTime = '''〔記者鍾智凱／台北報導〕第95屆奧斯卡今（13日）圓滿落幕，楊紫瓊以《媽的多重宇宙》拿下奧斯卡影后，寫下首位華裔影后的影史紀錄，結果沒想到今天資深女星楊繡惠也無預警冒出：「謝謝各位網友」。

去年楊紫瓊為雜誌《Town&Country》拍攝封面，走一個雍容華貴的造型，讓人看了頗驚艷，但沒想到被時尚專家認為撞臉楊繡惠，網友也非常驚詫表示「猛一看想說繡惠姊，瞬間質感攀升」，去年楊繡惠就曾說「來要看楊紫瓊的照過來」，搭配一系列激似楊紫瓊的宣傳照，今楊繡惠再度開腔「我要恭喜楊紫瓊小姐得獎　也要謝謝各位網友給我的恭喜」，全部湧入祝福讓她好笑又傻眼。

網友湧入楊繡惠臉書獻上恭喜。（翻攝自臉書）
楊繡惠神撞臉楊紫瓊，網友湧入粉絲頁面祝賀。（翻攝自臉書）
'''
```

```python
#開始使用 stanfordcorenlp
from stanfordcorenlp import StanfordCoreNLP

# 設定 CoreNLP 客戶端
nlp = StanfordCoreNLP('./stanford-corenlp-full-2018-02-27', lang='zh', timeout=30000)

word_result = nlp.word_tokenize(Title_LibTime + Content_LibTime)

# check result
print(word_result)
```

## 其他 Stanford Core NLP 的功能是否也都可以用？

```python
# POS : 詞性標記
pos_result = nlp.pos_tag(Title_LibTime + Content_LibTime)

print('POS 結果：' , pos_result)
```

```python
# parse : 描繪為樹狀形式
parse_result = nlp.parse( Title_LibTime + Content_LibTime )

print('Parse 結果：', parse_result)
```

```python
# dependency_parse : 解析為帶有樹狀結構的資訊
dp_result = nlp.dependency_parse( Title_LibTime + Content_LibTime )
print('帶有樹狀結構的解析：', dp_result)
```

```python
# 如何把這棵文字樹，畫得更漂亮？
```

```python
test_result = nlp.annotate(Content_LibTime)
print(test_result)
```
