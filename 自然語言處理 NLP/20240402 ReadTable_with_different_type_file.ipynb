# Read csv

```python
# Download CSV File
import os
Dataset_File = "Social_Network_Ads.csv"

if not os.path.isfile(Dataset_File):
  os.system("wget https://raw.githubusercontent.com/cnchi/datasets/master/" + Dataset_File)
```

```python
# Load CSV File
import pandas as pd

dataset = pd.read_csv(Dataset_File)
```

```python
# Print Dataset
print(dataset)
```
