# 將博客來中文書排行榜的原始內容抓下來，並且印出

```python
# 爬蟲設定
import requests
books_url = "https://www.books.com.tw/web/sys_tdrntb/books/?loc=subject_004"

# 開始抓取
books_url = requests.get(books_url, cookies={'over18': '1'}).text
print(books_url)
```

# 能將抓到的「書名、連結、作者」資訊，整理成 "title", "href", "author" 三個欄位，並以 JSON 格式存放後印出

```python
# 剖析抓到的原始資料
from bs4 import BeautifulSoup
import json

soup = BeautifulSoup(books_url, 'html.parser')
```

```python
book_data = []
# 逐一解析每本書
for book_item in soup.find_all('li', class_='item'):
    try:
      # 書名
      title_element = book_item.find('h4')
      title = title_element.text.strip()

    except :
      continue

    try:
      # 連結
      href = 'https://www.books.com.tw/web/' + book_item.find('a')['href']

    except :
      continue

    try:
      # 作者
      author_element = None
      for book_item in book_item.find_all('li'):
        if book_item.find('a'):
          author_element = book_item.find('a')
          break
      author_element = author_element.text.strip() if author_element else 'Unknown'

    except :
      continue


    # 將資訊存入文章列表
    book_data.append({
        "title": title,
        "href": href,
        "author": author_element
    })

# 將 book_data 串列，轉換成 JSON 格式印出
print(json.dumps(book_data, indent=2, ensure_ascii=False))
```

# 能夠將 JSON 格式存放的中文書排行榜，寫入 MongoDB 資料庫

```python
# 建立資料庫

# 將 JSON 字串轉換成 Python 字典 dict
# 印出字典內容驗證轉換成功
for item in book_data:
    print("title：", item["title"])
    print("web：", item["href"])
    print("author", item["author"])
    print("---------------------------------")
```

# 能將 MongoDB 資料庫內的所有資料讀出，並將它印在螢幕上

```python
# 將站點資訊存入 MongoDB 資料庫中
from pymongo import MongoClient
client = MongoClient()

# 建立一個 udn_news 資料庫與 books 資料集
book_element = client.books.sites

# 檢查資料是一筆還是很多筆
if isinstance(book_data, list): # json_dict 是否為 list
    book_element.insert_many(book_data)
else:
    book_element.insert_one(book_data)

# 將資料從 MongoDB 撈出來，驗證寫入成功
cursor = book_element.find()
for document in cursor:
    print(document)
```
