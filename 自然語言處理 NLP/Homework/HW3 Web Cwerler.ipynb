# 聯合新聞

## 記下該熱門新聞之 URL

```python
# 爬蟲設定
import requests
udn_url = "https://udn.com/rank/pv/2"

# 開始抓取
udn_text = requests.get(udn_url, cookies={'over18': '1'}).text
print(udn_text)
```

## 將抓取到的內容，整理成 JSON 格式

```python
# 剖析抓到的原始資料
from bs4 import BeautifulSoup
import json

soup = BeautifulSoup(udn_text, 'html.parser')
articles = []
```

```python
# 每篇文章都放在 <div class="r-ent"> ~ </div> 之間
for div in soup.find_all('div', class_='story-list__text'):
  try:
    # 標題放在 <div class="title"> ~ </div> 之間
    title = div.find('h2').text.strip()
    print(title)
    # 網址放在 <a> 裡面的 href 屬性內，且缺乏首頁網址
    href = div.find('a')['href']
    print(href)
    # 日期放在 <div class="date"> ~ </div> 之間
    time = div.find('time', class_='story-list__time').text.strip()
    print(time)
    # 將抓到的內容，存入 articles 串列中
    articles.append({
        'title': title,
        'href': udn_url,
        'time': time,
        'data source': '聯合新聞-' + 'https://udn.com/rank/pv/2'
    })
  except:
    continue
```

```python
# 將 articles 串列，轉換成 JSON 格式印出
print(json.dumps(articles, indent=2, ensure_ascii=False))
```

```python
# 建立資料庫

# 將 JSON 字串轉換成 Python 字典 dict
# 印出字典內容驗證轉換成功
for item in articles:
    print("title：", item["title"])
    print("web：", item["href"])
    print("time：", item["time"])
    print("資料來源：", item["data source"])
    print("---------------------------------")
```

```python
# 將站點資訊存入 MongoDB 資料庫中
from pymongo import MongoClient
client = MongoClient()

# 建立一個 udn_news 資料庫與 news 資料集
udn_news = client.news.sites

# 檢查資料是一筆還是很多筆
if isinstance(articles, list): # json_dict 是否為 list
    udn_news.insert_many(articles)
else:
    udn_news.insert_one(articles)

# 將資料從 MongoDB 撈出來，驗證寫入成功
cursor = udn_news.find()
for document in cursor:
    print(document)
```
