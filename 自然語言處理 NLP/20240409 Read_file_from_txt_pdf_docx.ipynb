# 讀取 TXT 檔案

```python
# 下載 TXT 檔案
import os
Dataset_File = "AI_News.txt"

if not os.path.isfile(Dataset_File):
  os.system("wget https://raw.githubusercontent.com/cnchi/datasets/master/" + Dataset_File)
```

```python
# 打開、讀取並顯示 AI_News.txt 中的內容
with open(Dataset_File, "r") as f:
  content = f.read()
  print(content)
```

# 讀取 PDF 檔案

```python
# 下載 PDF 檔案
import os
Dataset_File = "AI_News.pdf"

if not os.path.isfile(Dataset_File):
  os.system("wget https://raw.githubusercontent.com/cnchi/datasets/master/" + Dataset_File)
```

```python
# 安裝 PyMuPDF 套件
!pip install PyMuPDF
```

```python
# 打開、讀取並顯示 AI_News.pdf 中的內容
import fitz

with fitz.open(Dataset_File) as doc:
  content = ""
  for page in doc:
    content += page.get_text()
  print(content)
```

# 讀取 MS-Word 檔案

```python
# 下載 MS-Word 檔案
import os
Dataset_File = "AI_News.docx"

if not os.path.isfile(Dataset_File):
  os.system("wget https://raw.githubusercontent.com/cnchi/datasets/master/" + Dataset_File)
```

```python
# 安裝 python-docx 套件
!pip install python-docx
```

```python
# 打開、讀取並顯示 AI_News.pdf 中的內容
from docx import Document

doc = Document(Dataset_File)
content = ""
for para in doc.paragraphs:
    content += para.text
print(content)
```
